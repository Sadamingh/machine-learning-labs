# Heatmap and Correlations

```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; 
```

```python
sns.set()
np.random.seed(101)
```

```python
uniform_data = np.random.rand(3, 3)
uniform_data
```

Heatmap is basically mapping number to color...

```python
sns.heatmap(uniform_data, vmax=1.0, cmap="YlGnBu", annot=True)
```

## Get correlation coefficients with heatmap together

```python
x_data = np.linspace(0.0, 10.0, 100)
# noise
noise = np.random.randn(len(x_data))

# y = m*x + b + noise
y1 = (0.5 * x_data ) + 5 + noise

y2 = (-2 * x_data ) + 10 + noise
```

```python
# create pandas data frame
cols = ['x', 'y1', 'y2', 'noise']
df = pd.DataFrame(list(zip(x_data, y1, y2, noise)), columns=cols)
df.head()
```

```python
#correlation matrix
corrmat = df.corr()
#f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmin = -1, vmax = 1, annot=True, square=True);
```

```python
sns.set()
sns.pairplot(df, size = 1.5)
plt.show();
```

## Check the correlations in non-linear functions

```python
x_data = np.linspace(0.1, 10.0, 100)
y1 = x_data ** 2
y2 = x_data ** 2 + 10 * noise
y3 = x_data ** 2 - 10 * x_data + 30 
y4 = np.log(x_data) 

# create pandas data frame
cols = ['x', 'y1', 'y2', 'y3' , 'y4']
df = pd.DataFrame(list(zip(x_data, y1, y2, y3, y4)), columns=cols)
df.head()
```

```python
#correlation matrix
corrmat = df.corr()
#f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmin = -1, vmax = 1, annot=True, square=True);
```

```python
sns.set()
sns.pairplot(df, size = 1.5)
plt.show();
```
